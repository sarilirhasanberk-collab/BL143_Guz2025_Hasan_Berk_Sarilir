```python
# UYKU VE DERS ÇALIŞMA SAATLERİNİ ANALİZ EDEN BİR FONKSİYON KURUYORUZ.
def uyku_analiz():
    if uyku < 6:
        return "Yetersiz uyku"
    elif uyku <= 8:
        return "Normal uyku"
    else:
        return "Aşırı uyku"

def ders_analiz():
    if ders < 6:
        return "Normal ders"
    else:
        return "Yoğun ders"

#KULLANICIDAN VERİ ALMAK VE HATALI GİRİŞ YAPTIĞI ZAMAN TEKRAR SORMAK İÇİN DÖNGÜYE SOKUYORUZ.
while True:
    ders = input("Bugün kaç saat ders çalıştınız ? -> ")
    uyku = input("Bugünkü uyku sürenizi giriniz -> ")
    serbest = input("Bugünkü serbest takıldığınız zamanı giriniz ->")
    if not ders.isdigit() or not uyku.isdigit() or not serbest.isdigit():
        print("Lütfen sayısal veri giriniz.")
        continue
    #KULLANICIDAN ALINAN VERİLERİN SAYISAL VERİ OLDUĞUNU ÖĞRENDİK.
    #ŞİMDİ VERİLERİ İNTEGER DEĞER YAPIYORUZ.
    ders = int(ders)
    uyku = int(uyku)
    serbest = int(serbest)
    #KULLANICIDAN ALINAN VERİLERİN NEGATİF OLUP OLMADIĞINA BAKIYORUZ.
    if ders < 0 or uyku < 0  or serbest < 0:
        print("lütfen negatif değer girmeyiniz.")
        continue
    #KULLANICIDAN ALINAN VERİLERİN TOPLAMININ 24 SAAT OLUP OLMADIĞINA BAKIYORUZ.
    if ders + uyku + serbest != 24:
        print("Lütfen girdiğiniz verilerin tam olarak bir gün olmasına dikkat ediniz.")
        continue
    #TÜM KOŞULLAR UYGUN ÇIKTIĞINDA GÜNÜNÜ KONTROL ETMEYE BAŞLIYORUZ
    ders_durum = ders_analiz()
    uyku_durum = uyku_analiz()
    print(f"ders durumunuz = {ders_durum}.")
    print(f"uyku durumunuz = {uyku_durum}.")
    if uyku_durum == "Yetersiz uyku" and ders_durum == "Yoğun ders":
        print("Verimsiz bir gün geçirmişsiniz uykunuzu biraz daha almanızı öneririm.")
    else:
        print("verimli bir gün geçirmişsiniz.")
    break

```
